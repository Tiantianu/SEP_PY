```python
ppl2 = pd.read_table('/content/drive/MyDrive/ANTRA/people_2.txt')
ppl2.head()
```

```python
df= pd.concat([ppl1, ppl2])
df.head()
```

```python
#Name
df['FirstName']= df['FirstName'].str.capitalize()
df['LastName']= df['LastName'].str.capitalize()
```

```python
#Contact info
df['Phone']= df['Phone'].str.replace('-', '')
df['Email']= df['Email'].str.lower()
df["Address"] = df["Address"].apply(lambda x: x.translate({ord(c): None for c in "No.#"}))
```
